```python
import numpy as np
import pandas as pd
# Users/Pashlene/Desktop/pulsar.csv
data_intial = pd.read_csv('/Users/Pashlene/Desktop/pulsar.csv')
columns_name=data_intial.columns.tolist()
column_name=columns_name[:-1]
print(column_name)
data_intial.head()
```

```python
# Quantization Noise

perecentage=10
data1=data_intial

def add_noise(data,percentage):
    num_samples=data.shape[0]
    num_samples_to_select = int(percentage/100 * num_samples)
    selected_indices = np.random.choice(num_samples, num_samples_to_select, replace=False)
    
    if percentage < 100 and percentage >=0:
        for column in column_name:
            for index in selected_indices:
                data.loc[index, column] = int(data.loc[index, column].round())
                
        return True
        
    else :
        return False
    
add_noise(data1,10)
```

```python
data10=data_intial.copy()
data20=data_intial.copy()
data30=data_intial.copy()
data40=data_intial.copy()
data50=data_intial.copy()
```

```python
def generate_data(data_Name,percentage):
    add_noise(data_Name,percentage)
    return data_Name
```

```python
#datasets created with different pecentages of noise injected in dataset 
#d10 means 10% of dataset has injected quantized noise
d10=generate_data(data10,10,0,0.2)
d20=generate_data(data20,20,0,0.2)
d30=generate_data(data30,30,0,0.2)
d40=generate_data(data40,40,0,0.2)
d50=generate_data(data50,50,0,0.2)
```
